```python
for root, dirs, files in os.walk(BMBR_path):
    if len(files)==1:
        file=root+"/"+files[0]
        read_sequance=False
        sequence=""
        destiny_folder="data/"+files[0][:-6]
        try:
            os.system("mkdir "+destiny_folder)
        except:
            pass
        with open(file,"r") as f:
            for line in f:
                if "_Entity.Target_identifier" in line:
                    read_sequance=False
                    with open(destiny_folder+"/"+files[0][:-6]+".fasta", "w") as f:
                        f.write(sequence)
                if read_sequance:
                    if ";" not in line and len(line.split())>0:
                        sequence+=line.split()[0]
                if "_Entity.Polymer_seq_one_letter_code" in line and "_can" not in line:
                    read_sequance=True
```
