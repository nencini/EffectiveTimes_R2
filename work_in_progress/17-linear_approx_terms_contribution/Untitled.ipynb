```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import gc
gc.collect()
```

```python
gammaD=41.695*10**6; #r*s^(-1)*T^(-1)
gammaH=267.513*10**6;
gammaC=67.262*10**6;
gammaN=-27.166*10**6;

magn_field_MHz=850 # for values in MHz
magnetic_field=magn_field_MHz*2*np.pi/gammaH*10**6

mu = 4 * np.pi * 10**(-7) #magnetic constant of vacuum permeability
h_planck = 1.055 * 10**(-34); #reduced Planck constant
rN = 0.101 * 10**(-9); # average cubic length of N-H bond
d = 1 * (mu * gammaN * gammaH * h_planck) / (4 * np.pi * rN**3); # dipolar coupling constant


wh = gammaH * magnetic_field 
wn = gammaN * magnetic_field 

K1=(d**2 / 20)
K2=(wn * 160 * 10**(-6))**2 / 45
```

```python
# individual terms of Redfield for plotting purposes

def funcA(x):
    
    magn_field_MHz=850 # for values in MHz
    magnetic_field=magn_field_MHz*2*np.pi/gammaH*10**6

    mu = 4 * np.pi * 10**(-7) #magnetic constant of vacuum permeability
    h_planck = 1.055 * 10**(-34); #reduced Planck constant
    rN = 0.101 * 10**(-9); # average cubic length of N-H bond
    d = 1 * (mu * gammaN * gammaH * h_planck) / (4 * np.pi * rN**3); # dipolar coupling constant


    wh = gammaH * magnetic_field 
    wn = gammaN * magnetic_field 

    K1=(d**2 / 20)
    K2=(wn * 160 * 10**(-6))**2 / 45

    
    A=(wh-wn)**2
    B=wn**2
    C=wh**2
    D=(wh+wn)**2
    return K1*x/(1+A*x**2)
def funcB(x):
    magn_field_MHz=850 # for values in MHz
    magnetic_field=magn_field_MHz*2*np.pi/gammaH*10**6

    mu = 4 * np.pi * 10**(-7) #magnetic constant of vacuum permeability
    h_planck = 1.055 * 10**(-34); #reduced Planck constant
    rN = 0.101 * 10**(-9); # average cubic length of N-H bond
    d = 1 * (mu * gammaN * gammaH * h_planck) / (4 * np.pi * rN**3); # dipolar coupling constant


    wh = gammaH * magnetic_field 
    wn = gammaN * magnetic_field 

    K1=(d**2 / 20)
    K2=(wn * 160 * 10**(-6))**2 / 45
    
    A=(wh-wn)**2
    B=wn**2
    C=wh**2
    D=(wh+wn)**2
    return (3*K1+3*K2)*x/(1+B*x**2)
def funcC(x):
    magn_field_MHz=850 # for values in MHz
    magnetic_field=magn_field_MHz*2*np.pi/gammaH*10**6

    mu = 4 * np.pi * 10**(-7) #magnetic constant of vacuum permeability
    h_planck = 1.055 * 10**(-34); #reduced Planck constant
    rN = 0.101 * 10**(-9); # average cubic length of N-H bond
    d = 1 * (mu * gammaN * gammaH * h_planck) / (4 * np.pi * rN**3); # dipolar coupling constant


    wh = gammaH * magnetic_field 
    wn = gammaN * magnetic_field 

    K1=(d**2 / 20)
    K2=(wn * 160 * 10**(-6))**2 / 45
    
    A=(wh-wn)**2
    B=wn**2
    C=wh**2
    D=(wh+wn)**2
    return 6*K1*x/(1+C*x**2)
def funcD(x):
    magn_field_MHz=850 # for values in MHz
    magnetic_field=magn_field_MHz*2*np.pi/gammaH*10**6

    mu = 4 * np.pi * 10**(-7) #magnetic constant of vacuum permeability
    h_planck = 1.055 * 10**(-34); #reduced Planck constant
    rN = 0.101 * 10**(-9); # average cubic length of N-H bond
    d = 1 * (mu * gammaN * gammaH * h_planck) / (4 * np.pi * rN**3); # dipolar coupling constant


    wh = gammaH * magnetic_field 
    wn = gammaN * magnetic_field 

    K1=(d**2 / 20)
    K2=(wn * 160 * 10**(-6))**2 / 45
    
    A=(wh-wn)**2
    B=wn**2
    C=wh**2
    D=(wh+wn)**2
    return 6*K1*x/(1+D*x**2)

def funcLin(x):
    magn_field_MHz=850 # for values in MHz
    magnetic_field=magn_field_MHz*2*np.pi/gammaH*10**6

    mu = 4 * np.pi * 10**(-7) #magnetic constant of vacuum permeability
    h_planck = 1.055 * 10**(-34); #reduced Planck constant
    rN = 0.101 * 10**(-9); # average cubic length of N-H bond
    d = 1 * (mu * gammaN * gammaH * h_planck) / (4 * np.pi * rN**3); # dipolar coupling constant


    wh = gammaH * magnetic_field 
    wn = gammaN * magnetic_field 

    K1=(d**2 / 20)
    K2=(wn * 160 * 10**(-6))**2 / 45
    
    return (4*K1+4*K2)*x
```

```python
fig = plt.figure(figsize=(12,4))
outer = gridspec.GridSpec(1,3,wspace=0.25)
ax1 = plt.Subplot(fig,outer[0])
fig.add_subplot(ax1)
ax2 = plt.Subplot(fig,outer[1])
fig.add_subplot(ax2)
ax3 = plt.Subplot(fig,outer[2])
fig.add_subplot(ax3)


x=np.linspace(0,0.5*10**(-9))
ax1.plot(x*10**9,funcA(x),"-",label="K1*x/(1+Ax^2)",c="C0")
ax1.plot(1/(wh-wn)*10**9,funcA(1/(wh-wn)),"o",c="C0")
ax1.plot(x*10**9,funcB(x),"-",label="(3K1+3K2)*x/(1+Bx^2)",c="C4")
#ax1.plot(1/(-wn)*10**9,funcB(1/(-wn)),"o",c="C1")
ax1.plot(x*10**9,funcC(x),"-",label="6K1*x/(1+Cx^2)",c="C2")
ax1.plot(1/(wh)*10**9,funcC(1/(wh)),"o",c="C2")
ax1.plot(x*10**9,funcD(x),"-",label="6K1*x/(1+Dx^2)",c="C3")
ax1.plot(1/(wh+wn)*10**9,funcD(1/(wh+wn)),"o",c="C3")
ax1.plot(x*10**9,funcLin(x),"-",label="(4K1+4K2)*x",c="C1")
ax1.plot(x*10**9,funcLin(x)+funcA(x)+funcB(x)+funcC(x)+funcD(x),"-",c="C5",label="All the terms above")
#ax1.legend(prop={"size":5})
ax1.set_ylabel(r"y($\tau$) [s$^{-1}$]")
ax1.set_xlabel(r"$\tau$ [ns]")


x=np.linspace(0,2.5*10**(-9))
ax2.plot(x*10**9,funcA(x),"-",label="K1*x/(1+Ax^2)",c="C0")
ax2.plot(1/(wh-wn)*10**9,funcA(1/(wh-wn)),"o",c="C0")
ax2.plot(x*10**9,funcB(x),"-",label="(3K1+3K2)*x/(1+Bx^2)",c="C4")
ax2.plot(1/(-wn)*10**9,funcB(1/(-wn)),"o",c="C4")
ax2.plot(x*10**9,funcC(x),"-",label="6K1*x/(1+Cx^2)",c="C2")
ax2.plot(1/(wh)*10**9,funcC(1/(wh)),"o",c="C2")
ax2.plot(x*10**9,funcD(x),"-",label="6K1*x/(1+Dx^2)",c="C3")
ax2.plot(1/(wh+wn)*10**9,funcD(1/(wh+wn)),"o",c="C3")
ax2.plot(x*10**9,funcLin(x),"-",label="(4K1+4K2)*x",c="C1")
ax2.plot(x*10**9,funcLin(x)+funcA(x)+funcB(x)+funcC(x)+funcD(x),"-",c="C5",label="All the terms above")
#ax2.legend(prop={"size":5})
ax2.set_ylabel(r"y($\tau$) [s$^{-1}$]")
ax2.set_xlabel(r"$\tau$ [ns]")


x=np.linspace(0,20*10**(-9))
ax3.plot(x*10**9,funcLin(x),"-",label=r"(4K1+4K2)*$\tau$",c="C1")
ax3.plot(x*10**9,funcA(x),"-",label=r"K1*$\tau$/(1+A$\tau^2$)",c="C0")
ax3.plot(1/(wh-wn)*10**9,funcA(1/(wh-wn)),"o",c="C0")
ax3.plot(x*10**9,funcB(x),"-",label=r"(3K1+3K2)*$\tau$/(1+B$\tau^2$)",c="C4")
ax3.plot(1/(-wn)*10**9,funcB(1/(-wn)),"o",c="C4")
ax3.plot(x*10**9,funcC(x),"-",label=r"6K1*$\tau$/(1+C$\tau^2$)",c="C2")
ax3.plot(1/(wh)*10**9,funcC(1/(wh)),"o",c="C2")
ax3.plot(x*10**9,funcD(x),"-",label=r"6K1*$\tau$/(1+D$\tau^2$)",c="C3")
ax3.plot(1/(wh+wn)*10**9,funcD(1/(wh+wn)),"o",c="C3")

ax3.plot(x*10**9,funcLin(x)+funcA(x)+funcB(x)+funcC(x)+funcD(x),"-",c="C5",label="All the terms above")
ax3.legend(prop={"size":7.8})
ax3.set_ylabel(r"y($\tau$) [s$^{-1}$]")
ax3.set_xlabel(r"$\tau$ [ns]")
ax3.set_ylim(None,41)

plt.savefig("linear_approx.pdf")
```
